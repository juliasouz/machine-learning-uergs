```python
y=dados['Instant.Liking']
```

```python
x=dados.drop('Instant.Liking',axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold #o KFold para garantir o equilibrio dos dados
```

```python
kfold= StratifiedKFold(n_splits=5)
modelo=LogisticRegression(solver='liblinear')
resultado=cross_val_score(modelo,x,y,cv=kfold)
```

```python
resultado.mean()
```

```python
from sklearn.model_selection import GridSearchCV
valores_C=np.array([0.01,0.1,1,0.5,1,2,3,10,20,30,50,100])
regularizacao=['l1','l2']
valores_grid={'C':valores_C,"penalty":regularizacao}
grid_regress=GridSearchCV(estimator=modelo,param_grid=valores_grid,cv=5)
grid_regress.fit(x,y)
print(grid_regress.best_score_)
print(grid_regress.best_estimator_.C)
print(grid_regress.best_estimator_.penalty)
```

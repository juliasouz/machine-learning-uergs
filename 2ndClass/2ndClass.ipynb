```python
y = data['Instant.Liking']
```

```python
x = data.drop('Instant.Liking',axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold #o KFold para garantir o equilibrio dos dados
```

```python
kfold = StratifiedKFold(n_splits=5)
model = LogisticRegression(solver='liblinear')
result = cross_val_score(model,x,y,cv=kfold)
```

```python
result.mean()
```

```python
from sklearn.model_selection import GridSearchCV
values_C = np.array([0.01,0.1,1,0.5,1,2,3,10,20,30,50,100])
regularization = ['l1','l2']
grid_values = {'C':values_C,"penalty":regularization}
grid_regress = GridSearchCV(estimator=model,param_grid=grid_values,cv=5)
grid_regress.fit(x,y)
print(grid_regress.best_score_)
```

```python
print(grid_regress.best_estimator_.C)
```
